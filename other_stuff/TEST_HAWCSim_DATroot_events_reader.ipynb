```python
def Concatenate_Root_For_Training(path_to_DAT_root: str):

    dfs_smu = []
    dfs_nmu = []
    
    DAT = uproot.open(path_to_DAT_root)
    tree = DAT["XCDF"]
    df = tree.arrays(library="pd")

    evts_number = len(df["HAWCSim.Evt.Num"].unique())
    print(evts_number)

    for i in df["HAWCSim.Evt.Num"].unique():
        
        p = df[df["HAWCSim.Evt.Num"] == i] 
        p = dat.get_stations_info(p, "/home/grieco/SWGO_Soft/ML_multiPMT/tank_pos.txt")
        p = dat.get_shower_time(p)
        p = dat.get_dt_correction(p)
        p = dat.appy_time_correction(p)

        p_sm = dat.get_st_lvl_train_sm_df(p)
        p_nmu = dat.get_st_lvl_train_nmu_df(p)

        dfs_smu.append(p_sm)
        dfs_nmu.append(p_nmu)

    df_smu = pd.concat(dfs_smu, ignore_index=True)
    df_nmu = pd.concat(dfs_nmu, ignore_index=True)

    return df_smu, df_nmu
```

```python
df_traces_smu, df_traces_nmu  = Concatenate_Root_For_Training("root_files/DAT1405461.root")
```
