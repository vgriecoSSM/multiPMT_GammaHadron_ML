```python
from py_code import FPGA_feats_builder
import numpy as np
import pandas as pd
from pathlib import Path
import glob
import os
```

```python
traces_path_smu = Path("/home/grieco/SWGO_Soft/ML_multiPMT/dfs_traces/HAWCSim_array/train/sm/100160TeV_025deg_4FF_180R175h_Black/13petrain_13petest_500ns/")
traces_path_nmu = Path("/home/grieco/SWGO_Soft/ML_multiPMT/dfs_traces/HAWCSim_array/train/bkg/100160TeV_025deg_4FF_180R175h_Black/13petrain_13petest_500ns/")
save_path_smu = "/home/grieco/SWGO_Soft/ML_multiPMT/dfs_feats/FPGA_ML/sm/"
save_path_nmu = "/home/grieco/SWGO_Soft/ML_multiPMT/dfs_feats/FPGA_ML/bkg/"

for f in glob.glob(save_path_smu + "*.parquet"):
    os.remove(f)

for f in glob.glob(save_path_nmu + "*.parquet"):
    os.remove(f)
    
pe_min_sm = 13
pe_min_bkg = 13
max_time = 800

for parquet_file in traces_path_smu.glob("*.parquet"):
    name = parquet_file.stem   
    print(f"Processing file: {name}") 
    parquet_file_name = str(parquet_file)
    trace_path = parquet_file_name
    FPGA_feats_builder.build_df_feats_NOADC_mPMT(trace_path = trace_path, save_path = save_path_smu + name + ".parquet", mu_frac = 0, pe_threshold = pe_min_sm, pe_lim = np.inf, ch0_present = False, max_time = max_time)
print("All done! :))")
for parquet_file in traces_path_nmu.glob("*.parquet"):
    name = parquet_file.stem   
    print(f"Processing file: {name}") 
    parquet_file_name = str(parquet_file)
    trace_path = parquet_file_name
    FPGA_feats_builder.build_df_feats_NOADC_mPMT(trace_path = trace_path, save_path = save_path_nmu + name + ".parquet", mu_frac = 0, pe_threshold = pe_min_bkg, pe_lim = np.inf, ch0_present = False, max_time = max_time)
print("All done! :))")


feats_path_smu = Path("/home/grieco/SWGO_Soft/ML_multiPMT/dfs_feats/FPGA_ML/sm/")
feats_path_nmu = Path("/home/grieco/SWGO_Soft/ML_multiPMT/dfs_feats/FPGA_ML/bkg/")

feats_files = list(feats_path_smu.glob("*.parquet")) + list(feats_path_nmu.glob("*.parquet"))

dfs = [pd.read_parquet(f) for f in feats_files]
combined_df = pd.concat(dfs, ignore_index = True)

combined_df.to_parquet("/home/grieco/SWGO_Soft/ML_multiPMT/dfs_feats/FPGA_ML/sm_bkg/FPGA_df_mPMT_NOADC_13bkg13smu_pe.parquet")
```
