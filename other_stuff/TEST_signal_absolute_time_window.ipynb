```python
def get_stations_info(primary: pd.DataFrame, tank_pos_file_path : str, number_of_channels: int = 25501, channels_per_stations: int = 7):

    primary = primary.copy()

    PEs_ID = np.asarray(primary["HAWCSim.PE.PMTID"].iloc[0])
    Station = (PEs_ID // channels_per_stations).astype(int)

    max_station = number_of_channels // channels_per_stations
    Station = np.clip(Station, 0, max_station - 1)

    primary["PE.Station_ID"] = pd.Series([Station], index=primary.index, dtype=object)

    data_channels = np.loadtxt(fname = tank_pos_file_path)
    chID, xT, yT, rT = data_channels[:, 0], data_channels[:, 1], data_channels[:, 2], data_channels[:, 3]

    ch_IDposX_map = dict(zip(chID, xT))
    ch_IDposY_map = dict(zip(chID, yT))
    ch_IDposR_map = dict(zip(chID, rT))

    PE_ID_posX = np.array([ch_IDposX_map[pe] for pe in PEs_ID])
    PE_ID_posY = np.array([ch_IDposY_map[pe] for pe in PEs_ID])
    PE_ID_posR = np.array([ch_IDposR_map[pe] for pe in PEs_ID])

    primary["PE.Station_X"] = pd.Series([PE_ID_posX], index=primary.index, dtype=object)
    primary["PE.Station_Y"] = pd.Series([PE_ID_posY], index=primary.index, dtype=object)
    primary["PE.Station_R"] = pd.Series([PE_ID_posR], index=primary.index, dtype=object)      

    return primary
```

```python
last_minus_first_pe = []
pe_t_from_c = []


for i in df["HAWCSim.Evt.Num"].unique():

    p0 = df[df["HAWCSim.Evt.Num"] == i] 
    p0 = get_stations_info(p0, "tank_pos.txt")
    
    station_ID = np.asarray(p0["PE.Station_ID"].iloc[0])
    pe_times  = np.asarray(p0["HAWCSim.PE.Time"].iloc[0])
    PE_xT = np.asarray(p0["PE.Station_X"].iloc[0])
    PE_yT = np.asarray(p0["PE.Station_Y"].iloc[0])
    p_X = np.asarray(p0['HAWCSim.Evt.X'].iloc[0]) 
    p_Y = np.asarray(p0['HAWCSim.Evt.Y'].iloc[0])
    T_from_C = np.sqrt((PE_xT - p_X)**2 + (PE_yT - p_Y)**2)
    
    for station in np.unique(station_ID):
        mask = station_ID == station
        pe_times_ = pe_times[mask]
        PE_xT_ = PE_xT[mask]
        PE_yT_ = PE_yT[mask]
        T_from_C_ = np.sqrt((PE_xT_ - p_X)**2 + (PE_yT_ - p_Y)**2)
            
        last_pe_time = np.nanmax(pe_times_)
        first_pe_time = np.nanmin(pe_times_)
        diff_times = last_pe_time - first_pe_time
        if diff_times > 1:
            last_minus_first_pe.append(diff_times)
            pe_t_from_c.append(T_from_C_)
```

```python
print("Number of elements in the time difference = ", len(last_minus_first_pe))
print("Number of elements in the distance list = ", len(pe_t_from_c))
plt.figure(figsize = (7,6))
plt.hist(last_minus_first_pe, bins = 75, range = (0,150), density = True)
plt.xlabel("First PE time - Last PE time")
xticks = np.arange(0,150, 10)
plt.xticks(xticks, fontsize = 10)
#plt.xscale("log")
plt.xlim(-1, 150)
#plt.yscale("log")
plt.grid(alpha = 0.2)
plt.show()
```

```python
station_dist_from_core_meters = [float(pe_t_from_c[i][0])/100 for i in range(len(pe_t_from_c))]
plt.scatter(station_dist_from_core_meters, last_minus_first_pe, marker = ".", color = "red", alpha = 0.1)
#plt.yscale("log")
#plt.xscale("log")
plt.ylim(1e-2,200)
plt.xlabel("Distance from core [m]")
plt.ylabel("Time Difference Last minus First [ns]")
plt.show()
```
